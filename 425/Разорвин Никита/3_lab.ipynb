```python
import numpy as np

def Y(x):
    return(x**2-np.log(x+1))
def norma(x1,x2):
    return(abs(x1-x2))

def iter(pr,xpr):
    x_now=xpr-Y(xpr)/pr
    if (norma(x_now,xpr)>10e-4):
        iter(pr,x_now)
    else:
        print("x= ",x_now)
    return()
def proizv(x):
    return(2*x-1/(x+1))


x_start=2 #Далее переменная  не меняеется
iter(proizv(x_start),x_start)
```
